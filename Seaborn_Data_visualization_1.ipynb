# Visualizing Two Quantitative Variables
-------------------------------------------------------

1. We will use the ```relplot()``` by which we can draw **a scatter or line plot** at a time.

```
x,y,data
hue,hue_order,
col,row,col_order,col_wrap
size,style,
alpha
```
**NOTE** : if you want to subplots, use **col or row**

2. we will `relplot()` by setting `kind='line'`.

```
x,y,data,
kind,style,hue,
col,row
markers,dashes
```

## ScatterPlot

````
tips.head()
````

````
# Adjust further to add subplots based on family support
sns.relplot( x="total_bill", y="tip", 
            data=tips,
            kind="scatter")
# Show plot
plt.show()
````

````
# Adjust further to add subplots based on family support
sns.relplot( x="total_bill", y="tip", 
            data=tips,
            
            kind="scatter",
            
            col="day",
            col_order= [ 'Thur', 'Fri', 'Sat', 'Sun' ],
            col_wrap=2
          )
````

````
# Adjust further to add subplots based on family support
sns.relplot( x="total_bill", y="tip", 
            data=tips,
            
            kind="scatter",
            
            col="day",
            col_order= [ 'Thur', 'Fri', 'Sat', 'Sun' ],

            row='sex'
          )
````

````
# Adjust further to add subplots based on family support
sns.relplot( x="total_bill", y="tip", 
            data=tips,
            
            kind="scatter",
            
            style="sex",
            hue='sex',
            alpha=0.5
          )
````

````
# Adjust further to add subplots based on family support
sns.relplot( x="total_bill", y="tip", 
            data=tips,
            
            kind="scatter",
            
            size='size',
            hue='size'
          )
````

## Line Plots

````
data2=sns.load_dataset('mpg')
data2.head()
````

````
sns.relplot(x='model_year',y='mpg',data=data2,kind='line')

# Here the blurred lines are the confidence interval
# we can show the std. deviation also by ci='sd'
````

````
sns.relplot(x='model_year',y='mpg',data=data2,
            kind='line',ci=None,
            hue='origin',style='origin')
````

````
sns.relplot(x='model_year',y='mpg',data=data2,
            kind='line',ci=None,
            hue='origin',style='origin',
            dashes=False,markers=True)
````

````
# Subplots
sns.relplot(x='model_year',y='mpg',data=data2,
            kind='line',ci=None,
            col='origin',col_wrap=3,
            dashes=False,markers=True)
````

# Categorical Plots : sns.catplot()
----------------------------------------

```
1. Categorical scatterplots:

    stripplot() (with kind="strip"; the default)

    swarmplot() (with kind="swarm")

2. Categorical distribution plots:

    boxplot() (with kind="box")

    violinplot() (with kind="violin")

    boxenplot() (with kind="boxen")

3. Categorical estimate plots:

    pointplot() (with kind="point")

    barplot() (with kind="bar")

    countplot() (with kind="count")

```

## Stripplot and Swarmplot
-----------------------------------------------------------

There are type of scatter plots for categorical variables.

````
tips.head()
````

````
# Strip plot
sns.catplot(x='day',y='total_bill',hue='smoker',
            order=[ 'Thur', 'Fri', 'Sat', 'Sun' ], hue_order=['Yes','No'],
            data=tips, kind='strip')
````

````
# Swarm Plot
sns.catplot(x='day',y='total_bill',hue='smoker',
            order=[ 'Thur', 'Fri', 'Sat', 'Sun' ], hue_order=['Yes','No'],
            data=tips, kind='swarm')
````

## BarPlot and Countplot
--------------------------------------------------------

````
from numpy import mean

sns.catplot(x='day',y='total_bill',hue='smoker',
            order=[ 'Thur', 'Fri', 'Sat', 'Sun' ], hue_order=['Yes','No'],
            data=tips, kind='bar',
            estimator=mean, ci=95)

# By setting the ci as None, we can turn off the error bars

# Show plot
plt.show()
````

````
# Countplot

sns.catplot(x='sex',hue='smoker',
            order=[ 'Male','Female' ], hue_order=['Yes','No'],
            data=tips, kind='count')
````

## Point Plot

````
sns.catplot(kind='point',x='sex',y='total_bill',data=tips,
            hue='smoker',join=False,capsize=0.2)

# estimator : we can define which type of function we need the measure the central tendency
# ci : we can set the amount of confidence interval
````

# Box plot and Violin plot
------------------------------------------------------------------------

````
# Create a box plot with subgroups and omit the outliers
sns.catplot(x='day',y='total_bill',col='smoker',
            order=[ 'Thur', 'Fri', 'Sat', 'Sun' ], hue_order=['Yes','No'],
            data=tips, kind='box')

# Show plot
plt.show()

#sym for outliers   sym='' -> we can vanish the outliers 
#whis for whiskers  whis=2 -> 2 * IQR
````

````
# Create a box plot with subgroups and omit the outliers
sns.catplot(x='day',y='total_bill',hue='smoker',
            order=[ 'Thur', 'Fri', 'Sat', 'Sun' ], hue_order=['Yes','No'],
            data=tips, kind='violin')
````

# Customizing the plots
-------------------------------------------

1. we can change the style with ```sns.set_style```.
2. we can change the appearance with ```sns.set_context```
3. changing the color palette with ```sns.set_palette```
4. for *facet grid* type of figure,
    ```
    g= sns.relplot() 
    g.fig.suptitle('title',y=0.3) # Adding the title as whole
    g.set_titles('this is {col_name} ')
    g.set(xscale='log',xlabel='',ylabel='')
    ```

    for *axessubgrid* type of plots,
    ```
    g.set_title('')
    ```

````
# Change the color palette to "RdBu"
sns.set_style("whitegrid")
sns.set_palette("RdBu")

# Change the context to "poster",talk,notebook,paper
sns.set_context("poster")

g = sns.catplot(x="smoker" ,y="total_bill", data=tips,
                kind="box")

plt.xticks(rotation=90)
plt.show()
````
